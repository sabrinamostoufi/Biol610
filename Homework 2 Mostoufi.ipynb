```python
import numpy as np
from matplotlib import pyplot as plt
import scipy as sp
from scipy import stats as spStats
```

```python
# describe mean and standard deviation for each population size
population = np.genfromtxt("populations.txt")
populations = np.delete(population, 0,0)

year = populations[0:22, 0]
hare = populations[0:22, 1]
lynx = populations[0:22, 2]
carrot = populations[0:22, 3]

print("Mean population for hares is", hare.mean(), "+/-", np.std(hare))
print("Mean population for lynxes is", lynx.mean(), "+/-", np.std(lynx))
print("Mean population for carrots is", carrot.mean(), "+/-", carrot.std())
```

```python
#write a function to find which species was most abundant in each year

shape = np.shape(populations)

for x in range(shape[0]):
    year = populations[x:x+1,0]
    print(year)
    compare = populations[x:x+1,1:4]
    if (np.amax(compare)) == populations[x:x+1,1]:
        print("Hare")
    elif (np.amax(compare)) == populations[x:x+1,2]:
        print ("Lynx")
    else:
        print("Carrot")
```

```python
#plot population sizes of species as a function of time

X=populations
np.savetxt("populations1.txt",X)
Y=np.loadtxt("populations1.txt")
hares=plt.plot(Y[:,0],Y[:,1], 'b', label='Hares')
lynxes=plt.plot(Y[:,0],Y[:,2], 'r', label='Lynxes')
carrots=plt.plot(Y[:,0],Y[:,3], 'g', label='Carrots')
plt.title('Populations over time')
plt.legend(loc=1, ncol = 1)

plt.show
```
